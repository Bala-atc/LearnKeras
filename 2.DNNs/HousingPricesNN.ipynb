**First things first, import the necessary dependencies**

```
import keras
import numpy as np
import pandas as pd
```

**Next, import the data**

```
#If you want to import from Drive
from google.colab import drive
drive.mount('/content/drive')
df=pd.read_csv('drive/My Drive/Courses/LearnML/2.DNNs/HousingPrices.csv')
df.head()
```

**Now, we are going to set our x value to every attribute except for the sale price, and our y value to the sale price, since that will be what we are predicting**

```
X = df.drop(columns=['SalePrice'])
Y = df[['SalePrice']]
```

**Let's build the model**

```
model = keras.models.Sequential()

n_cols = X.shape[1]

model.add(keras.layers.Dense(10, activation='relu', input_shape=(n_cols,)))
model.add(keras.layers.Dense(10, activation='relu'))
model.add(keras.layers.Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

**When we fit the model here, we are going to do something different. Keras has a built in feature called EarlyStopping that will stop the model when no more progress is being made. This saves CPU power and helps prevent our model from overfitting.**

```
model.fit(X, Y, validation_split=0.2, epochs=30, callbacks=[keras.callbacks.EarlyStopping(patience=3)])
```

**We now want to make a prediction based on our model. To do this, we create an np.array with the different attributes of a house. The correct order is:**
1. YearBuilt
2. 2ndFlrSF
3. GrLivArea
4. FullBath	
5. HalfBath	
6. BedroomAbvGr	
7. TotRmsAbvGrd	
8. YrSold
**You can see the sample data we fed the network was the exact same as the first value in our dataset. The true value of this house was $208500, so you can see we were just a few thousand off.**

```
test_data = np.array([2003,	854,	1710,	2,	1,	3,	8,	2008])
print(model.predict(test_data.reshape(1,8), batch_size=1))
```

**Finally, if you want to save this model and use it for later, use the following code. You can see it saved to our colab environment**

```
model.save('saved_model.h5')
!ls
```

**To recover the model, use this code. You'll see the outputed number is the same as the one we predicted above, except for we restored the model from a file. By saving a model, it can more easily be deployed to production.**

```
old_model = keras.models.load_model('saved_model.h5')

test_data = np.array([2003,	854,	1710,	2,	1,	3,	8,	2008])
print(old_model.predict(test_data.reshape(1,8), batch_size=1))
```
